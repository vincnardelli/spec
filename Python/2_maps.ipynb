```python
import geopandas as gpd
import numpy as np
import os
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

BASE = "https://github.com/vincnardelli/spec/raw/main/data"
for f in ["tanzania.parquet", "kc_house.parquet", "kc_grid.parquet",
          "italian_provinces.parquet", "visium_hne_points.parquet"]:
    if not os.path.exists(f):
        !wget -q {BASE}/{f}
```

## Dati areali – Tanzania

```python
tanzania = gpd.read_parquet("tanzania.parquet")
tanzania.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
tanzania.plot(column="FEFRTRWTFR", legend=True, cmap="YlOrRd",
              legend_kwds={"label": "Fertility rate"}, ax=ax)
ax.set_title("Tanzania – Tasso di fecondità (DHS 2022)")
ax.set_axis_off()
plt.tight_layout()
plt.show()
```

## Dati puntuali – King County

```python
kc = gpd.read_parquet("kc_house.parquet")
kc.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

kc.plot(column="price", legend=True, cmap="YlOrRd", alpha=0.3,
        markersize=2, ax=axes[0])
axes[0].set_title("Price")
axes[0].set_axis_off()

kc["logprice"] = np.log(kc["price"])
kc.plot(column="logprice", legend=True, cmap="YlOrRd", alpha=0.3,
        markersize=2, ax=axes[1])
axes[1].set_title("Log(Price)")
axes[1].set_axis_off()

plt.tight_layout()
plt.show()
```

## Dati griglia – King County H3

```python
kc_grid = gpd.read_parquet("kc_grid.parquet")
kc_grid.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

kc_grid.plot(column="price", legend=True, cmap="YlOrRd", ax=axes[0])
axes[0].set_title("Price")
axes[0].set_axis_off()

kc_grid["logprice"] = np.log(kc_grid["price"])
kc_grid.plot(column="logprice", legend=True, cmap="YlOrRd", ax=axes[1])
axes[1].set_title("Log(Price)")
axes[1].set_axis_off()

plt.tight_layout()
plt.show()
```
